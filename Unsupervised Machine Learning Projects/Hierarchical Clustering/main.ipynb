# 1. Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import warnings

warnings.filterwarnings("ignore")
```

# 2. Creating an artificial dataset

```python
data = make_blobs(n_samples = 3000, n_features = 2, centers = 3, cluster_std = 1.3, random_state = 42)

data 
```

```python
dots = data[0]
dots
```

```python
import scipy.cluster.hierarchy as sch 
```

```python
plt.figure(figsize = (10, 5))
dendro_graph = sch.dendrogram(sch.linkage(dots, method = "ward"))
plt.show()
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(dots[:, 0], dots[:, 1])
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hierarchical_clustering = AgglomerativeClustering(n_clusters= 5, linkage= "ward")

y_hier_clus = hierarchical_clustering.fit_predict(dots)

y_hier_clus
```

```python
plt.figure(figsize = (10, 7))
plt.scatter(dots[y_hier_clus == 0, 0], dots[y_hier_clus == 0, 1], s = 150, c = "purple")
plt.scatter(dots[y_hier_clus == 1, 0], dots[y_hier_clus == 1, 1], s = 150, c = "green")
plt.scatter(dots[y_hier_clus == 2, 0], dots[y_hier_clus == 2, 1], s = 150, c = "orange")
plt.scatter(dots[y_hier_clus == 3, 0], dots[y_hier_clus == 3, 1], s = 150, c = "yellow")
plt.scatter(dots[y_hier_clus == 4, 0], dots[y_hier_clus == 4, 1], s = 150, c = "black")

plt.show()
```
