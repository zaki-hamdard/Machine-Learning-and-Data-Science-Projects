# 1. Importing the libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings("ignore")
```

# 2. Loading the Dataset

```python
df = pd.read_csv("./Iris.csv")
df.head()
```

# 3. Getting to know the Dataset

```python
df.info()
```

```python
df.drop(["Id"], axis = 1, inplace = True)
df.head()
```

```python
df.Species = df.Species.str.replace("Iris-", "")
df.head()
```

# 4. Performing Exporatory Data Analysis (EDA)

```python
sns.pairplot(df, hue = "Species", height = 2)
plt.show()
```

```python
df["Species_num"] = df["Species"]
df.head()
```

```python
df["Species_num"] = df["Species_num"].str.replace("setosa", "1")
df["Species_num"] = df["Species_num"].str.replace("versicolor", "2")
df["Species_num"] = df["Species_num"].str.replace("virginica", "3")

df.head()
```

```python
df.info()
```

```python
df["Species_num"] = pd.to_numeric(df["Species_num"], downcast = "integer")

df.info()
```

```python
df.drop(["Species"], axis = 1, inplace = True)
```

```python
df.head()
```

## Heatmap of Correlation Matrix

```python
plt.figure(figsize = (10, 6))
sns.heatmap(data = df.corr(), annot = True, cmap = "viridis")
plt.show()
```

```python
df.drop(["PetalLengthCm"], axis = 1, inplace = True)
```

```python
df.head()
```

# 5. Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

x = df.drop(["Species_num"], axis = 1)
y = df["Species_num"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

clf = RandomForestClassifier()

clf.fit(x_train, y_train)

y_preds = clf.predict(x_test)

print("Accuracy score: ", accuracy_score(y_test, y_preds) * 100)
```

# 6. Model Evaluation

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, classification_report
cv_score = cross_val_score(estimator = clf, X = x_train, y = y_train, cv = 10)
cv_score.mean()
```

```python
cnf_matrix = confusion_matrix(y_test, y_preds)

sns.heatmap(data = cnf_matrix, annot = True, cmap = "YlGnBu")
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.show()
```

```python
print(classification_report(y_test, y_preds))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

features = list(df.columns[:-1])
targets = ["setosa", "versicolor", "virginica"]

plt.figure(figsize = (20, 20))

tree = tree.plot_tree(clf.estimators_[0], filled = True, feature_names = features, class_names = targets)
```

# 7. Hyper Parameter Optimization

```python
model2 = RandomForestClassifier()

params = {
    "n_estimators": [50, 100, 300],
    "max_depth": [3, 5, 7],
    "max_features": [2, 4, 6],
    "min_samples_split": [2, 4, 6]
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_cv = GridSearchCV(estimator = model2, cv = 5, param_grid = params, n_jobs = -1)

grid_cv.fit(x_train, y_train)
```

```python
grid_cv.best_params_
```

```python
tuned_model = RandomForestClassifier(max_depth = 3, max_features = 2, min_samples_split = 2, n_estimators = 50)

tuned_model.fit(x_train, y_train)

y_preds_tuned = tuned_model.predict(x_test)

print("Accuracy: ", accuracy_score(y_test, y_preds_tuned))
```

```python
print(confusion_matrix(y_test, y_preds_tuned))
```

```python
sns.heatmap(confusion_matrix(y_test, y_preds_tuned), annot = True, cmap = "YlGnBu")
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.show()
```

### Accuracy after hyper parameter optimization

```python
print(classification_report(y_test, y_preds_tuned))
```

##### So, now we can see that our model has indeed optimized and reached accuracy of 100%.
